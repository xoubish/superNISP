```python
import astropy.io.fits as fits

# Path to your FITS file
fits_path = 'COSMOS2020_FARMER_R1_v2.2_p3.fits'

# Load the FITS file
with fits.open(fits_path) as hdul:
    data = hdul[1].data  # Adjust the HDU index if your data is not in the first extension
```

```python
# Extract RA and DEC
ra = data['ALPHA_J2000']  # Adjust column name as necessary
dec = data['DELTA_J2000']  # Adjust column name as necessary

# Path to the output DS9 region file
region_file_path = 'output.reg'

# Write to the region file
with open(region_file_path, 'w') as file:
    file.write('# Region file format: DS9 version 4.1\n')
    file.write('global color=green dashlist=8 3 width=1 font="helvetica 10 normal roman" select=1 highlite=1 dash=0 fixed=0 edit=1 move=1 delete=1 include=1 source=1\n')
    file.write('fk5\n')  # Coordinate system

    for r, d in zip(ra, dec):
        file.write(f"circle({r}, {d}, 0.1') # color=red\n")  # 0.1 arcmin radius, adjust as needed

print("DS9 region file created:", region_file_path)
```
